
If weekend sales are higher, running weekend flash sales could be a smart strategy.

```python
weekend_sales = df[df['Day_Name'].isin(['Saturday', 'Sunday'])]['Purchase_Amount'].sum()
weekday_sales = df[~df['Day_Name'].isin(['Saturday', 'Sunday'])]['Purchase_Amount'].sum()

print("Weekend Sales:", weekend_sales)
print("Weekday Sales:", weekday_sales)
```

# Customer Purchase Behavior Analysis 📊download.jpeg




📌 Project Overview
This project analyzes customer purchase behavior using a dataset of 1000 customers. The goal is to uncover valuable insights that help businesses optimize marketing strategies, improve customer retention, and increase sales.


🔍 Key Insights & Business Impact
1️⃣ High-Value Customers (Top Spenders 💰)
Top spenders are middle-aged professionals with high-income levels.

Business Action: Target high-value customers with personalized offers & exclusive deals.

2️⃣ Customer Loyalty & Retention 🔄
Customers with higher Brand Loyalty Scores spend more.

Loyalty program members tend to spend significantly more than non-members.

Business Action: Expand customer loyalty programs & reward repeat buyers.

3️⃣ Best Time for Marketing Campaigns 🕒
Most purchases happen between 6 PM - 9 PM & on weekends.

Business Action: Schedule ads & promotions during peak shopping hours.

4️⃣ Discounts & Pricing Strategy 🎯
Discount-sensitive customers spend less per purchase but buy more frequently.

Business Action: Offer targeted discounts to encourage higher total spending.

5️⃣ Shopping Channel & Device Preference 📱💻
Online shopping is more preferred than Offline purchases.

Business Action: Invest in digital marketing & enhance online shopping experience.
